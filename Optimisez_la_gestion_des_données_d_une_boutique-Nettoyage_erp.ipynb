# Recherche des erreurs de saisie

```python
# Recherche erreurs de saisie 'instock'

instock = dataerp[dataerp['stock_status'] == 'instock']
errorinstock = instock[instock['stock_quantity'] == 0]

errorinstock
```

```python
# Recherche erreurs de saisie 'outofstock'

outofstock = dataerp[dataerp['stock_status'] == 'outofstock']
erroroutofstock = outofstock[outofstock['stock_quantity'] != 0]

erroroutofstock
```

- 1 produit n'ayant pas le bon 'stock_quantity' ou 'stock_status'

# Correction des erreurs de saisie

```python
# Correction des produits en stock 

dataerp['stock_status'] = np.where(dataerp['stock_quantity']== 0, 'outofstock', 'instock')
```
