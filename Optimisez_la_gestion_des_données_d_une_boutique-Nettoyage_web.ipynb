```python
import pandas as pd
import numpy as np
```

```python
dataweb = pd.read_csv('../Data/raw_data/web.csv')
```

```python
dataweb.info()
```

```python
dataweb
```

- La première colonne du DataFrame ne porte pas le bon nom
- Dans la première colone nous erreurs lexical ( bon-cadeau-25-euros & 13127-1)
- Nous avons plusieurs colones avec des valeurs manquantes ou des valeurs null
- Le format des dates est à changer

# Correction du nom de la colonne

```python
# Renommer la colonne 'sku' en 'id_web'

dataweb = dataweb.rename(columns={'sku':'id_web'})
```

```python
dataweb.info()
```

# Vérification et traitement des valeurs null (NaN)

```python
# Recherche des valeurs null (NaN)

print(dataweb.isnull().sum())
```

- Il y a 83 lignes de valeurs null (NaN)
- Il y a 4 colonnes ayant que des valeurs null (NaN)
- Il y a 3 colonnes ayant beaucoup de valeurs null (NaN)

```python
# Remplacement des 0 par de NaN

dataweb = dataweb.replace(0, np.nan)
print(len(dataweb.index),'lignes avant la suppression des lignes NaN')
dataweb.head()
```

- On remplace tout les 0 par des NaN afin de supprimer toute les lignes contenant que des valeurs null

```python
# Suppression des lignes de NaN

dataweb = dataweb.dropna(how = 'all')
print(len(dataweb.index),'lignes aprés la suppression des lignes NaN')
```

```python
# Suppression des 'id_web' NaN

dataweb[dataweb['id_web'].isnull()]
```

- On voit qu'on a 2 produits sans id_web

```python
# Suppression des produits n'ayant pas de 'id_web'

dataweb = dataweb.dropna(subset=['id_web'])
```

# Modification du DataFrame

- Afin d'avoir des données comprehensible et utile, on garde seulement les colones importantes.

```python
# Changement du DataFrame en gardant que les colonnes pertinantes

dataweb = dataweb[['id_web','total_sales','post_date','post_title','post_excerpt','post_name','post_type']]
dataweb
```

# Vérification et traitement des doublons

```python
# Recherche des doublons

dataweb.loc[dataweb['id_web'].duplicated(keep=False),:]
```

- Ici nous avons des doublons pour chaque produits, mais ils ont un post_type différent (product & attachement)

Le "product" est le produit alors que le "attachement" est un objet attaché au produit comme par exxemple une photo

```python
# Suppréssion des doublons

dataweb = dataweb.drop_duplicates(subset=['id_web'])
dataweb
```
