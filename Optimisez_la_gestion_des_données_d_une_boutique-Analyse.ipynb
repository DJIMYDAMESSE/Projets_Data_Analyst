```python
# Import des bibliothèques

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import des fichier données nettoyer

dataerp = pd.read_csv("Data/processed_data/erp_V2.csv")
dataliaison = pd.read_csv('Data/processed_data/liaison_V2.csv')
dataweb = pd.read_csv('Data/processed_data/web_V2.csv')
```

```python
# Premier Merge entre les données du fichier erp et les données du fichier liaison

dataLE = pd.merge(dataliaison, dataerp, how="inner", on=["product_id"])
dataLE
```

```python
# Vérirification des valeurs null après la fusion des données

print(dataLE.isnull().sum())
```

**Il n'y a aucune valeur null !**

```python
# Deuxième Merge entre le premier Merge et les données du fichier web

data = pd.merge(dataLE, dataweb, how="inner", on=["id_web"])
data
```

```python
# Vérirification des valeurs null après la fusion des données

print(data.isnull().sum())
```

**Il n'y a aucune valeur null !**

```python
# Plus d'informations sur nos données

data.info()
```

# Chiffre d'affaire par produits

```python
# Chiffre d'affaire = prix du produit * nombre de produit vendu

data = data.assign(CA = data['price']*data['total_sales'])
data[['product_id','id_web','price','total_sales','post_title','CA']]
```

# Chiffre d'affaire total

```python
total = sum(data['CA'])
print("Nous avons un chiffre d'affaire total de :",total,"€")
```

# Analyse des prix des produits

```python
data.boxplot(column='price', vert=False)
plt.title('Analyse des prix et valeurs abérantes')
plt.xlabel('Prix')
plt.show()
```

```python
moyenne = np.mean(data['price'])
mediane = np.median(data['price'])

Q1 = round(np.percentile(data['price'], 25))
Q2 = round(np.percentile(data['price'], 50))
Q3 = round(np.percentile(data['price'], 75))

IQ = Q3 - Q1

A = min(data['price'])
B = Q3+(1.5*IQ)

print("Moyenne des prix : ",round(moyenne,2))
print("Médiane : ",mediane)
print("Q1 = ",Q1)
print("Q3 = ",Q3)
print("IQ = ",IQ)
print("min = ",A)
print("max = ",B)
```

- 75% des produits ont un prix compris entre 5,2€ et 42€
- 50% des produits ont un prix compris entre 14€ et 42€
- Les Outlier sont supèrieur à 84€

Attention nous sommes ici sur le marché du vin de haute qualité, donc ces valeurs dites aberrantes peuvent refléter la qualité du vin.

# Liste des Outlier

```python
outlier = data[data['price']>84]
print(len(outlier), "produits ayant un prix concidérer comme valeur extreme")
outlier
```
